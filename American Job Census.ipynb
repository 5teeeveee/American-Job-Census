# Import Data and Remove NA Values

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
%matplotlib inline

#Remove NA values
recent_grads = pd.read_csv('recent-grads.csv')
print ('Original number of rows:', recent_grads.shape[0])
recent_grads2 = recent_grads.dropna(axis=0, how ='any')
print('New number of rows:', recent_grads2.shape[0])

#Create column of %men in school
recent_grads2.loc[:,'ShareMen']= recent_grads2.loc[:,'Men'].divide(recent_grads2.loc[:,'Total'])
```

# Custom made scatter matrix

```python
fig = plt.figure(figsize= (8,8))
fig.subplots_adjust(wspace =0, hspace = 0 )
fig.suptitle('College Major Female Enrollment vs Unemployment Rate Scatter Matrix', fontsize = 15)

sub1 = fig.add_subplot(221)
sub1.hist(recent_grads2['ShareWomen'], range = (recent_grads2['ShareWomen'].min(), recent_grads2['ShareWomen'].max()))
xaxis = sub1.get_xaxis()
yaxis = sub1.get_yaxis()
xaxis.set_visible(False)
sub1.set_ylim(0,30)
sub1.set_ylabel('ShareWomen')
sub1.set_yticklabels([0, 5, 10, 15, 20, 25, 30]) 

sub2 = fig.add_subplot(222)
sub2.scatter(recent_grads2['Unemployment_rate'],recent_grads2['ShareWomen'])
xaxis = sub2.get_xaxis()
yaxis = sub2.get_yaxis()
yaxis.set_visible(False)
xaxis .set_visible(False)
sub2.set_ylim(0,1)
sub2.set_yticklabels([0.00, 0.05, 0.10, 0.15]) 

sub3 = fig.add_subplot(223)
sub3.scatter(recent_grads2['ShareWomen'], recent_grads2['Unemployment_rate'])
xaxis = sub3.get_xaxis()
yaxis = sub3.get_yaxis()
sub3.set_xlabel('ShareWomen')
sub3.set_ylabel('Unemployment_rate')
sub3.set_xlim(0,1)
sub3.set_ylim(0,0.2)
sub3.set_xticklabels([0.0, 0.2, 0.4, 0.6, 0.8], rotation=90) 

sub4 = fig.add_subplot(224)
sub4.hist(recent_grads2['Unemployment_rate'], range = (recent_grads2['Unemployment_rate'].min(), recent_grads2['Unemployment_rate'].max()))
xaxis = sub4.get_xaxis()
yaxis = sub4.get_yaxis()
yaxis.set_visible(False)
sub4.set_xlabel('Unemployment_rate')
sub4.set_xlim(0,0.2)
sub4.set_xticklabels([0.00, 0.05, 0.10, 0.15, 0.20], rotation=90)

```

Initial Analysis: No clear correlation between women percentage and unemployment rate.

```python
arts_df = recent_grads2[recent_grads2['Major_category']=='Arts']
locs = numpy.arange(arts_df.shape[0])

fig2 = plt.figure(figsize=(8,8))
fig2.suptitle('Gender Ratio of Arts Majors', fontsize = 15)
sub1 = fig2.add_subplot(111)
sub1.bar(left = locs, height = arts_df['ShareMen'], width = 0.35)
sub1.set_xticklabels(arts_df['Major'].tolist(), rotation = 90)

offset_locs = locs + 0.35
bar_2 =sub1.bar(left = offset_locs, height = arts_df['ShareWomen'], width = 0.35, color = 'green')

sub1.set_xticks(offset_locs)
sub1.legend(['ShareMen', 'ShareWomen'], loc = 'upper left')
```

Initial Analysis: Women hold a >65% percentage in visual, performing, studio, drama, theater, commercial and graphic design arts. Music is split roughly 1:1 between males and females. The only male majority arts majors are miscellaneous fine arts and film/video/photographics arts.
